```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# hago una petición mediante request en formato texto:
res = requests.get("https://deliveroo.es/es/restaurants/madrid/legazpi").text

# análisis sintáctico como html mediante BeautifulSoup:
soup = BeautifulSoup(res, 'html.parser')

# prettify lo coloca de forma más legible. Print:
print(soup.prettify())
```

```python
#
listadoRest = soup.select('div[class^="HomeFeedUICard"] ul li[class^="HomeFeedUILines"] span p')
```

```python
def resName(lista):
    return[x.text for x in lista]
```

```python
names = resName(listadoRest)

letras = [len(e) for e in names]
letras
```

```python
df = pd.DataFrame(zip(names, letras), columns=("nombre", "letras"))

df
```

```python
sorted = df.sort_values("letras")
sorted
```
