```python
import pandas as pd
```

```python
df = pd.read_csv("input/data.csv")
display(df.head())
```

## Compruebo el porcentaje de datos nulos por columna:

```python
vNulo = (100*df.isnull().sum()/len(df)).round(1)
PN = pd.DataFrame(vNulo).rename(columns = {0 : '% de nulos'})
display(PN)
```

## Elimino columnas cuyos valores nulos sean más del 80%

```python
for col in df.columns:
    if vNulo[col] >= 80:
        df.drop(col, axis=1, inplace=True)
```

## También elimino columnas cuyos valores no me interesan:

```python
lista = ['Class of Orbit',
         'Type of Orbit',
         'Longitude of GEO (degrees)',
         'Perigee (km)',
         'Apogee (km)',
         'Eccentricity',
         'Inclination (degrees)',
         'Launch Mass (kg.)',
         'Dry Mass (kg.)',
         'Power (watts)',
         'COSPAR Number',
         'Comments',
         'Source Used for Orbital Data',
         'Source',
         'Source.1',
         'Source.2',
         'Detailed Purpose',
         'Expected Lifetime (yrs.)']

for e in lista:
    df.drop(e, axis=1, inplace=True)
```

## Y renombro algunas de ellas para facilitar la legibilidad:

```python
df.columns=(['Official Name', 'Country/Org of UN Registry',
       'Country of Operator', 'Operator', 'Users', 'Purpose',
       'Period (minutes)', 'Date of Launch', 'Contractor',
       'Country of Contractor', 'Launch Site', 'Launch Vehicle',
       'NORAD Number'])
```

## Añado una columna 'Year' con la fecha de lanzamiento:

```python
import re

def extractYear(x):
    year = re.findall(r'[\d*]{4}', x)
    return year[0]
    
df["Year"] = (df["Date of Launch"]).apply(extractYear)
df
```

## Compruebo la cantidad de valores nulos por columna

```python
def v_nulo(df):
    
        vNulo = df.isnull().sum()
        vNulo_percent = 100 * df.isnull().sum() / len(df)
        vNulo_table = pd.concat([vNulo_percent, vNulo], axis=1)
        vNulo_table_ren_columns = vNulo_table.rename(columns = {0 : '%', 1 : 'Valores Nulos'})
        
        vNulo_table_ren_columns = vNulo_table_ren_columns[vNulo_table_ren_columns.iloc[:,1] != 0].sort_values(
        '%', ascending=False).round(1)
        
        print ("Tu dataframe tiene " + str(df.shape[1]) + " columns.\n"      
            "Hay " + str(vNulo_table_ren_columns.shape[0]) +
              " columnas con valores nulos.")
        
        return vNulo_table_ren_columns
```

```python
v_nulo(df)
```
